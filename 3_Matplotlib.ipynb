# Sample Figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([2, 5, 7, 4, 7, 0, 3, 1, 9, 2])
```

```python
plt.plot(a)
plt.title("My first figure")
plt.xlabel("My x-axis")
plt.ylabel("My y-axis")
plt.show()
```

# Subfigures

```python
fig, ax = plt.subplots(2, 2)
print(ax.shape)
ax[0, 0].plot(np.arange(6))
ax[0, 1].plot(np.arange(6, 0, -1))
ax[1, 0].plot((-1)**np.arange(6))
ax[1, 1].plot((-1)**np.arange(1, 7))
```

- Matlabby Way

```python
plt.subplot(2, 2, 1)
plt.plot(np.arange(6))
plt.subplot(2, 2, 2)
plt.plot(np.arange(6, 0, -1))
plt.subplot(2, 2, 3)
plt.plot((-1)**np.arange(6))
plt.subplot(2,2, 4)
plt.plot((-1)**np.arange(1, 7))
```
